```python
import pandas as pd

#Read in Planning Application file
#For reproducing code replace with your own file path
file = "/content/drive/MyDrive/project/planningApplications.csv"

#Converted file to a dataframe
df = pd.read_csv(file)

#Inspected columns names
for col in df.columns:
    print(col)
```

```python
#Inspected sample of data to understand necessary processing, transformations and feature engineering
df.sample(10, random_state=1)
```

```python
#Created list of columns to keep
keep_cols = [
    "OBJECTID",
    "Planning Authority",
    "Application Number",
    "Development Address",
    "Development Postcode",
    "Application Status",
    "Application Type",
    "Decision",
    "Number of Residential Units",
    "One-Off House",
    "Shape__Area"
]

#Removed all other rows from datafraem
df = df[keep_cols]


```

```python
#Inspected unique values for 'Number of Residential Units'
print(df["Number of Residential Units"].unique())

#Converted all values to numeric type, setting incompatible data types to null
df["Number of Residential Units"] = pd.to_numeric(df["Number of Residential Units"], errors="coerce")

#Reduced data to only applications for one-off houses
df = df[df["Number of Residential Units"] == 1]

#Inspected new reduced data
df.sample(10, random_state=1)



```

```python
#Inspected values in One-Off House Column
#Kept all remaining entries as there wasn't any contradictory values (E.g. 'No', '0')
#Nulls a result of differing fields on applications in different planning jurisdictions
print(df["One-Off House"].unique())
```

```python
#Inspected unique values for Application Types
print(df["Application Type"].unique())

#Applied domain knowledge based on research to reduce the data down to outright planning permission applications only
#Applications for outline permission, extension permission, retention permission etc. were removed
valid_app_types = [
    'PERMISSION',
    'Permission',
    'APPROVAL',
    'Approval',
    'PERMISSION CONSEQUENT',
    'Permission Consequent'
]

df = df[df["Application Type"].isin(valid_app_types)]

#Inspected sample of new reduced data
df.sample(10, random_state=1)
```

```python
#Inspected unique values of Application Status column
print(df["Application Status"].unique())

#Again, applied domain knowledge based on research to reduce data to only applications that were fully completed and finalised
#Removed applications that were incomplete, withdrawn, appealed etc.
valid_statuses = [
    'APPLICATION FINALISED',
    'DECISION MADE',
    'Decision Made',
    'Decision',
    'Final Grant',
    'Final Grant Review',
    'Decision Following A.I.',
    'Decision Notice Issued',
    'Planners Report'
]
df = df[df["Application Status"].isin(valid_statuses)]

#Inspected new reduced data
df.sample(10, random_state=1)
```

```python
#Inspected unique values for Decision column
print(df["Decision"].unique())

#Once again, applied domain knowledge based on research to assign each decision type to a binary value of either Accepted (1) or Rejected (0)
#Conditional cases were deemed to be accepted as ultimately the development is being allowed
decision_map = {
    'CONDITIONAL': 1,
    'UNCONDITIONAL': 1,
    'Granted (Conditional)': 1,
    'GRANT PERMISSION': 1,
    'GRANT APPROVAL': 1,
    'GRANT PERMISSION FOLLOWING OUTLINE': 1,
    'GRANT PERMISSION AND RETENTION PERMISSION': 1,
    'REFUSED': 0,
    'Refused': 0,
    'REFUSE PERMISSION': 0,
    'REFUSE APPROVAL': 0,
    'SPLIT DECISION(PERMISSION & REFUSAL)': 0
}
df['Decision Binary'] = df['Decision'].map(decision_map)

#Removed any instances where the verdict was unclear
df = df.dropna(subset=['Decision Binary'])

#Converted column to int data type
df['Decision Binary'] = df['Decision Binary'].astype(int)

#Inspected new reduced data
df.sample(10, random_state=1)
```

```python
#Inspected value counts of Decision column to understand class imbalances
#Heavy oversampling of positive case requires treatment for machine learning
#However, this 87% acceptance rate is the same as is reported by the CSO
df['Decision Binary'].value_counts()
```

```python
import re

#Applied preprocessing to addresses before being passed to Photon API to be converted to GPS coordinates
#Aim was to add missing county and country context to addresses to make it easier for API to convert accurately

#Inspected unique values in Planning Application column
df['Planning Authority'].unique()

#Removed council suffixes from Planning Authority values
df['pa_clean'] = df['Planning Authority'].str.extract(
    r'^(.*?)\s*(?:City and County Council|City Council|County Council)', expand=False
).str.strip()

#Replaced South Dublin with Dublin as the county itself is more likely to be contained in addresses
df['pa_clean'] = df['pa_clean'].replace({'South Dublin': 'Dublin'})

#For addresses that didn't contain their county already, added ", {county}, Ireland"
#For all other addresses, just added ", Ireland"
def build_full_address(dev_address, pa_clean):
    if pd.isna(dev_address):
        dev_address = ""
    address = re.sub(r'\s+', ' ', str(dev_address)).strip()
    address = re.sub(r',+', ',', address)
    if pa_clean.lower() not in address.lower() and pa_clean != "":
        address = f"{address}, {pa_clean}"
    address = f"{address}, Ireland"
    address = re.sub(r',+', ',', address).strip().strip(',')
    return address

#Applied method to dataframe, adding new column
df['full_address'] = df.apply(lambda row: build_full_address(row['Development Address'], row['pa_clean']), axis=1)

#Inspected new augmented data
df.sample(10, random_state=1)
```

```python
#Saved reduced, processed planning data to csv
#Ready to be passed to geocoding Python script
df.to_csv("/content/drive/MyDrive/project/planningAppsCleaned.csv", index=False)
```
